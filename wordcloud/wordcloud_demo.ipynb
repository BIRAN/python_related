```
# This exmaple is from https://github.com/amueller/word_cloud
# you can find more information there.

import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = "Love"

x, y = np.ogrid[:300, :300]

mask = (x - 150) ** 2 + (y - 150) ** 2 > 150 ** 2
mask = 255 * mask.astype(int)


wc = WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

text = open('/content/qixi.txt','r').read()
yeezy_coloring = np.array(Image.open("/content/img01.jpg"))


wc = WordCloud(background_color="white", max_words=2000, mask=yeezy_coloring,
               max_font_size=40, random_state=3, repeat = True)

wc.generate(text)
image_colors = ImageColorGenerator(yeezy_coloring)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure()
```

```
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.figure()
```
